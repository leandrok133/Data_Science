## DISTIBUIÇÕES DISCRETAS

Distribuições discretas modelam eventos que podem ser contados em números inteiros, ou seja, valores discretos.

**Apliação:**

* Aplicam-se a situações onde as variáveis são discretas, como contagem de eventos.

**Principais Distribuições:**

* Distribuição Binomial;
* Distribuição Geométrica;
* Distribuição Binomial Negativa;
* Distribuição de Poisson;


### DISTRIBUIÇÃO BINOMIAL

```r
# Exemplo: Definindo como sucesso o cliente comprar, e supondo que a probabilidade de sucesso é 50%.
# Ao passar 10 clientes em nossa loja, qual a probabilidade de realizarmos 2 vendas?
#Ou seja, queremos encontrar a probabilidade de dois sucessos, em dez tentativas. Cuja probabilidade de sucesso
# em cada tentativa é 50%
dbinom (x = 2, size = 10, prob = 0.5)
#Onde:
# x é o número de sucessos,
# size é o número de tentativas,
# prob é a probabilidade de sucesso em cada tentativa
```

```r
# A função a seguir gera quantidades aleatórias de sucesso oriundos de uma quantidade (size) de tentativas dada a probabilidade
#(prob) de sucesso.
# É útil para realizar experimentos. Podemos simular qual a frequencia esperada de vendas a cada dez clientes ?

#Ainda mantendo a probabilidade de sucesso (cliente comprar) de 50%
va_binomial <- rbinom(n = 30, size=10, prob=0.5)
#Onde:
# n é a quantidade de vezes que o experimento deve ser repetido
# size é o número de tentativas a cada experimento
# prob é o número de sucesso em cada uma das tentativas
hist(va_binomial) # A maior barra no histograma representa a quantidade esperada de vendas
```

```r
#Ajustando o parametro n para 1000 e plotando o histograma, podemos observar que a distribuição binomial se aproxima da normal
hist(rbinom(n = 1000, size = 10, prob = 0.5))
```

```r
# Podemos também querer a probabilidade de que até dois clientes comprem.
#Ao invés de saber a probabilidade de exatos dois comprarem.
#A probabilidade de até dois clientes comprarem é:
#(probabilidade de nenhum cliente comprar) + (probabilidade de um cliente comprar) + probabilidade de dois cliente comprarem)
#Formalizando: P(X<=2) = P(X=0) + P(X=1) + P(X=2)
pbinom(q = 2,size = 10, prob = 0.5)
#A probabilidade de que até dois clientes comprem ao entrarem dez clientes, é de 5,48%
```

## DISTRIBUIÇÃO GEOMÉTRICA

```r
#Exemplo: Definindo como sucesso o cliente comprar, e supondo que a probabilidade de sucesso é 50%.
#Qual a probabilidade da primeira venda ocorrer quando o quinto cliente entrar na loja?
dgeom(x = 5-1, prob = 0.5)
#Onde:
# x é o número de fracaços até o primeiro acerto
# prob é a probabilidade de sucessos
```

### DISTRIBUIÇÃO POISSON

```r
# Exemplo: Uma loja recebe em média, 6 (𝝺) clientes por minuto. Qual a probabilidade de que 5(x) clientes
#entrem em um minuto?
dpois(x= 5,lambda = 6)
61
#Onde:
# x é a quantidade a ser testada
# lambda é a taxa média de ocorrêcia do evento em um determinado período de intervalo de tempo ou espaço
# Podemos utilizar a mesma funcao para obter a probabilidade de entrar um cliente, dois clientes... quinze clientes
va_poison <- dpois(x = 1:15, lambda = 6)
plot(va_poison)
# Observe que os valores se distribuiem simetricamente en tormo de seis, use acontece porque o paramentro
#lambda é a média (e também o desvio padrão) da distribuição de Poisson
# Também podemos obter a probabilidade acumulada de até 5 clientes entrarem na loja em um minuto
#Formalizando, queremos: P(X<=5)
va_poison <- ppois(1:5, lambda = 6)
plot(va_poison)
```

## INTERVALO DE CONFIANÇA

O intervalo de confiança é um conceito fundamental na estatística que fornece uma faixa de valores dentro da qual acredita-se que um parâmetro populacional, como a média ou a proporção, esteja contido com uma determinada probabilidade. Ele é usado para quantificar a incerteza associada a estimativas amostrais e oferece uma medida de quão confiáveis são essas estimativas.

**Principais Pontos**

* Estimativas Amostrais: Quando trabalhamos com amostras de uma população, nossas estimativas (por exemplo, média amostral) podem variar de uma amostra para outra. O intervalo de confiança fornece um intervalo que abrange uma ampla gama de possíveis valores para o parâmetro populacional, levando em consideração essa variação.

* Nível de Confiança: O intervalo de confiança é acompanhado por um nível de confiança, geralmente expresso em porcentagem (por exemplo, 95% de confiança). Isso indica a probabilidade de que o intervalo contenha o verdadeiro valor do parâmetro. Quanto maior o nível de confiança, mais amplo será o intervalo.

* Importância: O intervalo de confiança é importante porque ajuda a interpretar a precisão das estimativas amostrais. Se tivermos um intervalo estreito com alta confiança, podemos ter maior confiança em nossas conclusões. Por outro lado, se o intervalo for amplo, nossa estimativa é menos precisa.

* Uso Generalizado: Ele permite que os pesquisadores comuniquem a incerteza associada a seus resultados de uma maneira compreensível e transparente.

* Tamanho da Amostra: O tamanho da amostra desempenha um papel crucial na largura do intervalo de confiança. Quanto maior a amostra, mais estreito será o intervalo, dada a mesma confiança.

## Intervalo de confiança para média amostral pela distribuição Normal Padrão

```r
# Obter o intervalo de confiança para uma variável cuja média = 30, desvio padrão = 7,31 e n = 30
#Temos que definir o nível de confiança do nosso intervalo.
#Podemos obter o valor do quantil para o nível de confiança desejado com a função qnorm()
#O quantil na distribuição normal padrão para 95% de confiança
ic <- 0.95
alfa <- 1-ic

qnorm(1-(alfa/2))
```

```r
#Vamos armazenar os valores em objetos
media <- 30
desvio_padrao_populacional <- 7.31
n <- 30
quantil_95 <- qnorm(0.975)
#Aplicando a fórmula para o intervalo de confiança, temos:
Limite_Superior <- n + quantil_95*(desvio_padrao_populacional/sqrt(n))
Limite_Inferior <- n - quantil_95*(desvio_padrao_populacional/sqrt(n))
paste("Com 95% de confiança, podemos afirmar que a média varia entre",Limite_Inferior," e ",Limite_Superior)


#Podemos também visualizar onde está nossa média da amostra em relação aos limites do IC
# Seta uma distribuição normal padrão
va_normal <- rnorm(n = n, mean = media,sd = desvio_padrao_populacional)
plot(density(va_normal)) #Plota curva de densidade
abline(v = Limite_Inferior,col = 'red') #Gera uma linha com o Limite_Inferior
abline(v = Limite_Superior,col = 'red') #Gera uma linha com o Limite_Superior
abline(v = mean(va_normal),col = 'blue') #Gera uma linha com o a média da amostra
```

### Intervalo de confiança para a média amostral pela distribuição t de Student

```r
#A teoria nos diz para utilizar a distribuição t de Student quando não soubermos o desvio padrão populacional.
#Vamos assumir que o desvio padrão que temos é obtido a partir da amostra
#Vamos armazenar os valores em objetos
media <- 30
desvio_padrao_amostral <- 7.31
n <- 30
quantil_95_t <- qt(0.975,df = n-1)
#Aplicando a fórmula para o intervalo de confiança da distribuição t de Student, temos:
Limite_Superior_t <- 30+quantil_95_t*(desvio_padrao_amostral/sqrt(n))
Limite_Inferior_t <- 30-quantil_95_t*(desvio_padrao_amostral/sqrt(n))
paste("Com 95% de confiança, podemos afirmar que a média varia entre",Limite_Inferior_t," e ",Limite_Superior_t)


#Podemos também visualizar onde está nossa média da amostra em relação aos limites do IC
# Seta uma distribuição t
va_t <- rt(n = n, df = n -1, ncp = media)
plot(density(va_t)) #Plota curva de densidade
abline(v = Limite_Superior_t,col = 'red') #Gera uma linha com o Limite_Inferior
abline(v = Limite_Inferior_t,col = 'red') #Gera uma linha com o Limite_Superior
abline(v = mean(va_t),col = 'blue') #Gera uma linha com o a média da amostra
```

```r
#Podemos descobrir o intervalo de confiança de um conjunto de dados de forma mais direta
#Vamos gerar com o comando rnorm(), uma amostra com n= 30 observações
#apartir de uma distribuição normal com média 30 e desvio padrão 7,31.
va <- rnorm(n = 30, mean = 30, sd = 7.31)

#Calculando o intervalo de 95% de confiança com a distribuição t de Student
#com a funçao t.test()
IC <-t.test(va, conf.level = 0.95)
IC$conf.int
#Obs.: Por se tratar de uma amostra, a média, desvio padrão e intervalo de confiança
#serão ligeiramente diferentes da distribuição normal que os gerou e, por isso,
#não coincidirão com o exemplo anterior.
#ou seja, a média e desvio padrão de va são diferentes de 30 e 7,31
```

### Intervalo de confiança para a proporção

```r
# Exemplo: 138 de n = 500 clientes realizaram a devolução do produto
#Vamos armazenar os valores em objetos
devolucoes <- 138
n <- 500
quantil_95 <-qnorm(0.975)
proporcao_devolucoes <- devolucoes/n
#Aplicando a fórmula para o intervalo de confiança de uma proporção:
Limite_Superior_prop <- proporcao_devolucoes + quantil_95 * sqrt(proporcao_devolucoes*(1-proporcao_devolucoes)/n)
Limite_Inferior_prop <- proporcao_devolucoes - quantil_95 * sqrt(proporcao_devolucoes*(1-proporcao_devolucoes)/n)
paste("Com 95% de confiança, podemos afirmar que a proporção varia entre",Limite_Inferior_prop," e ",Limite_Superior_prop)
```

```r
#Podemos obter o intervalo de confiança para proporção mais fácil pela função prop.test()
IC_proporcao <- prop.test(x = 138, n = 500, conf.level = 0.95)
IC_proporcao$conf.int
```

### Intervalo de confiança para média via Bootstrap

```r
#O método usa o Teorema Central do Limite (TCL), que a firma:
#Mesmo que a variável aleatória não obedeça uma distribuição normal,
#a distibuição das médias amostrais obece

#Vamos gerar uma va seguindo uma distribuição qui-quadrado
va <- rchisq(n = 100, df = 3)
#Observe o quão assimétrica é a va
hist(va)
```

```r
#Inicializa variaveis
medias <- c() #Essa variável é um vetor para armazenar a média de cada subamostra bootstrap
R <- 1000 #Numero de subamostras extraídas para gerar a distribuição amostral de médias
#bootstrap
for (i in 1:R) {
  #Realiza uma subamostragem aleatória com reposição da va
  reamostra <- sample(va, size = 50, replace = T)
  #Armazena a média da subamostra
  medias[i] <- mean(reamostra)
}
#Distribuicao das médias das subamostras (distribuição amostral da média da va)
hist(medias)
```

```r
#Observe que mesmo a variável original não seguindo uma distribuição normal, o Teorema Central do Limite
#nos garante que a distribuição das médias será normal se n é suficientemente grande
#A partir das médias geradas, precisamos achar dois valores, o que corta a cauda inferior
#e o que corta a cauda superior da distribuição. Lembrando que ela é simétrica
#Caso o intervalo desejado seja de 95% de confiança, temos que ordenar essa distribuição
#do menor valor para o maior e achar o valor que deixará 2,5% dos dados para trás e o
#valor que deixará 97,5% para trás
(1-0.95)/2
1-(1-0.95)/2
#Visualize o intervalo de confiança via bootstrap
quantile( medias, probs = c(0.025,0.975))
```

```r
#Vamos realizar mais um experimento
#Geraremos uma va com média = 30 e desvio padrão amostral =7.31 e n = 30
va <- rnorm(n = 30, mean = 30, sd = 7.31)
#Iremos calcular o intervalo de confiança usando o Bootstrap e também com
#a distribuiçãot de Student e compararemos os resultados.
#Inicializa variavel para armazenar as médias de cada subamostra
medias <- c()
R <- 10000 #Numero de subamostras extraídas para gerar a distribuição amostral de médias
#bootstrap
for (i in 1:R) {
  #Realiza uma subamostragem aleatória com reposição da va
  reamostra <- sample(va, size = 20, replace = T)
  #Armazena a média da subamostra
  medias[i] <- mean(reamostra)

}
#Distribuicao das médias das subamostras (distribuição amostral da média da va)
hist(medias)
#Limites inferior e superior do intervalo pelo bootstrap
quantile( medias, probs = c(0.025,0.975))
#Limites inferior e superior do intervalo via t de Student
IC<-t.test(va, conf.level = 0.95)
IC$conf.int
```

## TESTE DE HIPÓTESE

Um teste de hipótese é um procedimento estatístico usado para tomar decisões com base em dados amostrais. Ele é projetado para determinar se uma afirmação ou hipótese específica sobre uma população é estatisticamente significativa ou não.

>**Principais Pontos**

**Hipótese Nula (H0) e Hipótese Alternativa (H1):** No teste de hipótese, duas hipóteses são formuladas.
* A hipótese nula (H0) representa uma afirmação inicial, muitas vezes alegando que não há efeito ou diferença significativa.
* A hipótese alternativa (H1) representa a afirmação oposta, sugerindo que existe um efeito ou diferença significativa.

**Teste Estatístico:** Com base nos dados coletados, um teste estatístico é aplicado para calcular um valor estatístico, como o valor-p. O valor-p indica a probabilidade de observar os resultados se a hipótese nula for verdadeira.

**Tomada de Decisão:** Com base no valor-p e no nível de significância escolhido (geralmente 0,05), uma decisão é tomada. Se o valor-p for menor que o nível de significância, rejeita-se a hipótese nula em favor da hipótese alternativa. Se o valor-p for maior que o nível de significância, a hipótese nula não é rejeitada.

>**Importância**

Os testes de hipótese são fundamentais na pesquisa científica e na tomada de decisões baseadas em dados. Eles permitem que os pesquisadores avaliem se suas hipóteses são suportadas pelas evidências amostrais.

Através dos testes de hipótese, é possível realizar experimentos controlados, estudar relações de causa e efeito, validar teorias científicas e avaliar o impacto de intervenções em diversos campos, como medicina, economia, psicologia, entre outros.

### Avaliando a normalidade de uma variável aleatória
Realizar um teste de normalidade é uma etapa fundamental na análise estatística, pois influencia a escolha de técnicas estatísticas apropriadas, a interpretação dos resultados e a validade das conclusões. No entanto, é importante observar que a normalidade não é sempre necessária, e existem métodos estatísticos robustos que podem ser aplicados a dados não normalmente distribuídos quando apropriado.

```r
set.seed(10)
#Gera v.a. que segue distribuição normal com n = 70, média = 40 e desvio padrão = 8
va_normal <- rnorm(n = 70, mean = 25, sd = 8)
#Gera v.a. que segue uma distribuição F (não normal) com n = 15, 2 graus de liberdade no numerados e 10 graus de liberdade no denominador
va_nao_normal <- rf(n =15, df1 =2, df2 = 10)
#Visualize o histograma das variáveis geradas
#Observe como os dados se distribuem em torno do valor médio na va normal
hist(va_normal)
```

```r
#Observe como os dados não se distribuem em torno de um valor médio exibindo padrão assimétrico
hist(va_nao_normal)
```

```r
# Visualize o QQ-Plot
# Observe como os pontos de dados seguem a linha reta qq norm da va normal
qqnorm(va_normal)
qqline(va_normal) #Este comando é para adicionar a linha
```

```r
# Observe como os pontos de dados não seguem a linha reta na va não normal
qqnorm(va_nao_normal)
qqline(va_nao_normal) #Este comando é para adicionar a linha
```

```r
# Vamos aplicar o teste de hipóteses Shapiro Wilk. O teste funciona sob as hipóteses
# H0: A variável segue uma distribuição normal
# H1: A variável não segue uma distribuição normal
# Fixe um nível de significância alfa e analise o p valor (p-value) do Shapiro Wilk
#Se o p-value for menor que alfa a hipótese nula deve ser rejeitada
shapiro.test(va_normal)
shapiro.test(va_nao_normal)
```

### Teste t para diferença de médias (duas amostras independentes)

O teste t para diferença de médias, também conhecido como teste t de duas amostras independentes, é uma técnica estatística utilizada para determinar se há evidências estatísticas de diferença significativa entre as médias de duas amostras independentes (ou grupos).

>**Aplicações Comuns**

* Teste de Eficácia de Tratamento: Usado em estudos clínicos para avaliar se um tratamento (por exemplo, um medicamento) é eficaz em comparação com um grupo de controle.

* Análise de Desempenho de Produtos: Aplicado na indústria para comparar o desempenho de produtos ou processos sob diferentes condições.

* Estudos de Mercado: Usado para analisar se existe uma diferença significativa nas preferências ou comportamentos de dois grupos de consumidores.

>**Requisitos para Aplicação**

* Independência das Amostras: As duas amostras devem ser independentes uma da outra, o que significa que a observação em uma amostra não está relacionada de forma alguma com a observação na outra amostra.

* Distribuição Normal ou Amostras Grandes: Idealmente, as amostras devem seguir uma distribuição normal. No entanto, quando o tamanho das amostras é grande (geralmente acima de 30), o teste t é robusto o suficiente para lidar com desvios da normalidade.

* Homogeneidade das Variâncias: É importante que as variâncias das duas amostras sejam aproximadamente iguais. A violação desse requisito pode ser tratada com versões modificadas do teste t, como o teste t de Welch.

* Hipóteses e Nível de Significância: O pesquisador deve definir claramente as hipóteses nula e alternativa e escolher um nível de significância (geralmente 0,05) antes de realizar o teste.

```r
#Iremos testar se:
# H0: As vendas na posição A são iguais as Vendas na Posição B
# H1: As vendas na posição A são diferentes das vendas na posição B
rm(list = ls()) #Limpa objetos da memória do R
mu1 <- 150.1 #Armazena as média de vendas na posição A
mu2 <- 182.1 #Armazena as média de vendas na posição B
s1 <- 17 #Armazena o desvio padrão das vendas na posição A
s2 <- 19.2 #Armazena o desvio padrão das vendas na posição B
n1 <- 25 #Armazena a quantidade observações registradas para de vendas na posição A
n2 <- 30 #Armazena a quantidade observações registradas para de vendas na posição B

#Calcula nossa estatística de teste. Que é o t calculado
t <- (mu1 - mu2) / sqrt( s1^2/n1 + s2^2/n2)
paste('t calculado:', t) #Visualize o valor de t calculado
#Calcula os graus de liberdade da estatística de teste
gl <- (s1^2/n1 + s2^2/n2)^2 /( (s1^2/n1)^2 / (n1-1) + (s2^2/n2)^2 / (n2-1) )
paste('grau de liberdade:', gl) #Visualize a quantidade de graus de liberdade
#Obtem o quantil (t crítico) para uma distribuição t com gl graus de liberdade. A um alfa de 5%
quantil <- qt(0.975,df = gl)
paste('t calculao:', quantil) #Visualize o t crítico
#Esse é o aspecto de uma distribuição t com n=53 observações e com n - 1 graus de liberdade
plot(density(rt(n = 53,df = gl)),xlim = c(-7,7))
#Observe onde estão os valores críticos que acabamos de encontrar
abline(v = quantil,col = 'blue',lwd = 2)
abline(v = -quantil,col = 'blue',lwd = 2)
abline(v = t, col = 'red')# Observe como o tcalculado é muito menor que o tcrítico. Está na região de rejeição
```

```r
#Obtendo o valor p
#P(Tcalculado > Tcritico)
paste('p calculado:', 2*pt(q = t, df = gl))
#Agora vamos realizar o mesmo teste de hipótese utilizando a função nativa do R t.test()
vendas_A <- rnorm(n= 25, mean = 150.1, sd = 17)
vendas_B <- rnorm(n = 30, mean = 182.1, sd = 19.2)
#Observe no output desta função, que ela já nos da tudo pronto, t calculado e valor p
t.test(vendas_A,vendas_B, alternative = 'two.sided')
#Esse é o aspecto de uma distribuição t com n observações e com n - 1 graus de liberdade
n <- 5
plot(density(rt(n = n,df = n-1)))
#Altere o valor de n de 5 em 5 observe que a medida que os graus de liberdade aumenta a distribuição se aproxima da normal. Como os valores são gerados aleatoriamente

#poderemos ter curvas diferentes para um mesmo valor de n, mas a medida que n cresce o comportamento simétrico tende a estabilizar.
```

### Teste t para diferença de médias (duas amostras dependentes)

O teste t para diferença de médias de duas amostras dependentes, também conhecido como teste t pareado, é uma técnica estatística utilizada para determinar se há evidências estatísticas de diferença significativa entre as médias de duas amostras que estão relacionadas de alguma forma. Nesse teste, as observações nas duas amostras estão emparelhadas, ou seja, cada observação em uma amostra está relacionada a uma observação na outra amostra.

>**Aplicações Comuns**

* Teste de Eficácia de Tratamento: Usado em estudos clínicos para avaliar se um tratamento (por exemplo, um novo medicamento) é eficaz em comparação com o mesmo grupo de pacientes antes e depois do tratamento.

* Teste de Desempenho de Produto: Aplicado na indústria para avaliar se uma modificação em um produto resulta em um desempenho significativamente diferente em relação à versão anterior.

* Avaliação de Treinamento: Usado para verificar se a eficácia de um programa de treinamento é estatisticamente significativa ao comparar o desempenho dos participantes antes e depois do treinamento.

>**Requisitos para Aplicação**

* Amostras Dependentes: As duas amostras devem ser dependentes, ou seja, cada observação em uma amostra deve estar relacionada a uma observação na outra amostra. Isso pode ocorrer quando as observações são medidas na mesma unidade, indivíduo ou entidade antes e depois de uma intervenção.

* Distribuição Normal ou Amostras Grandes: Idealmente, as amostras devem seguir uma distribuição normal. No entanto, quando o tamanho das amostras é grande (geralmente acima de 30), o teste t é robusto o suficiente para lidar com desvios da normalidade.

* Homogeneidade das Variâncias: É importante que as variâncias das diferenças entre os pares de observações sejam aproximadamente iguais.

* Hipóteses e Nível de Significância: O pesquisador deve definir claramente as hipóteses nula e alternativa e escolher um nível de significância (geralmente 0,05) antes de realizar o teste.

```r
# H0: O peso médio após a dieta é igual ao peso médio antes da dieta
# H1: O peso médio após a dieta é menor do que o peso médio antes da dieta
rm(list = ls()) #Limpa memória do R
#Iremos utilizar uma biblioteca adicional para gerar valores aleatórios que sigam uma distribuição normal entre um intervalo de valor para simular os pesos
#A biblioteca chama 'truncnorm'. Basta instalar com o comando abaixo install.packages().
#Uma vez instalada não há mais necessidade de instalar novamente. Basta carregar com o comando library()
install.packages('truncnorm')
library(truncnorm)
set.seed(100)
```

```r
#Gera uma amostra aleatória, seguindo uma distribuição normal cujo valor mínimo é 100 e o valor máximo é 140.
#O valor de n=20, média = 123 e desvio padrão 18
#Com essa v.a. iremos simular os pesos dos indivíduos antes da dieta
antes_da_dieta <- rtruncnorm(n=20, a=100, b=140, mean=123, sd=18)
#Gera uma amostra aleatória, seguindo uma distribuição normal cujo valor mínimo é 110 e o valor máximo é 130.
#O valor de n=20, média = 110 e desvio padrão 28
#Com essa v.a. iremos simular os pesos dos indivíduos após a dieta
depois_da_dieta <- rtruncnorm(n=20, a=110, b=130, mean=110, sd=28)
#Calcula a diferença depois da dieta e antes da dieta, para cada indivíduo
diferenca <- depois_da_dieta-antes_da_dieta
#Visualiza a distribuicao da diferença de pesos
hist(diferenca)
shapiro.test(diferenca) #Avalie a normalidade da distribuição da diferença
```

```r

#Aplica test t com os seguintes argumentos
t.test(depois_da_dieta,antes_da_dieta,
  paired = TRUE, #Pareado

  alternative = "less", #Unilateral a esquerda
  conf.level = 0.9 #90 porcento de confiança
)
#O comando t.test() acima nos da tudo que precisamos para executar e concluir o teste. Mas a título de conhecimento, podemos realizar o teste passo a passo
```

```r
#Calcula a média das diferenças
media <- mean(diferenca)
#Desvio padrão das diferenças
desvio_padrao <- sd(diferenca)
#Quantidade de indivíduos
n <- 20
#Obtem o t calculado
t_calculado <- media / (desvio_padrao/sqrt(n))
#Obtem o valor p para o t calculado com n - 1 graus de liberdade.
pt(q = t_calculado, df = n-1)
#Podemos também obter o t crítico para uma distribuição t com 19 (n-1=20-1) graus de liberdade ao nível de confiança de 90%
tcrítico_teste_t_pareado <- -qt(p = 0.9, df = 19) #Devido ao teste ser unilateral a esquerda a distribuição t ser simétrica, nossa estatística de teste será negativa
```

```r
#Observe que o t calculado é maior que o t critico. Como estamos em um teste unilateral a esquerda o t calculado
#estará fora da região de rejeição caso seja maior que o t crítico
t_calculado < tcrítico_teste_t_pareado
```

### Teste Qui-Quadrado para associação entre variáveis categóricas

O teste Qui-Quadrado é uma técnica estatística utilizada para avaliar a associação ou independência entre duas variáveis categóricas em um conjunto de dados. Ele é especialmente útil quando se deseja determinar se existe uma relação significativa entre as categorias das variáveis em uma tabela de contingência.

>**Aplicações Comuns**

* Pesquisas de Opinião e Marketing: Utilizado para entender se existe uma relação entre variáveis categóricas, como idade e preferências de produtos, sexo e hábitos de consumo, etc.

* Ciências Sociais: Aplicado para investigar associações entre características demográficas, como educação e renda, e opiniões políticas, filiações religiosas, etc.

* Ciências da Saúde: Usado para estudar associações entre fatores de risco (por exemplo, tabagismo) e doenças (por exemplo, câncer de pulmão) em grupos demográficos específicos.

>**Requisitos para Aplicação**

* Variáveis Categóricas: O teste Qui-Quadrado é adequado para variáveis categóricas, ou seja, aquelas que representam categorias ou grupos, como sim/não, alto/baixo, jovem/idoso.

* Tabela de Contingência: É necessário criar uma tabela de contingência que organize os dados em relação às duas variáveis categóricas. A tabela deve mostrar a frequência de ocorrência de combinações entre as categorias das duas variáveis.

* Hipóteses e Nível de Significância: O pesquisador deve estabelecer claramente as hipóteses nula e alternativa antes do teste. O nível de significância (geralmente 0,05) também deve ser definido.

* Independência das Observações: As observações devem ser independentes uma das outras. Isso significa que a inclusão de um indivíduo ou caso em uma categoria não deve afetar a inclusão de outro indivíduo em outra categoria.

>**Interpretação dos Resultados**

* Valor Qui-Quadrado (χ²): O teste Qui-Quadrado gera um valor de estatística de teste (χ²) que é comparado a uma distribuição Qui-Quadrado sob a hipótese nula de independência. Quanto maior o valor χ², maior a evidência contra a independência.

* Graus de Liberdade (df): Os graus de liberdade são determinados pela estrutura da tabela de contingência e são usados para encontrar os valores críticos da distribuição Qui-Quadrado.

* Valor-p: O valor-p é calculado com base no valor χ² e indica a probabilidade de obter os resultados observados, supondo-se que as variáveis sejam independentes. Um valor-p baixo (geralmente menor que 0,05) sugere uma associação significativa.

### ANOVA
O teste de hipótese com Análise de Variância, ou ANOVA, é uma técnica estatística usada para determinar se existem diferenças significativas entre as médias de três ou mais grupos independentes. É uma ferramenta poderosa para investigar se, em vez de comparar apenas dois grupos, existem diferenças estatísticas entre múltiplos grupos.

>**Aplicações Comuns**

* Estudos Clínicos: Usado para comparar os efeitos de diferentes tratamentos ou medicamentos em várias populações.

* Testes de Qualidade: Aplicado na indústria para verificar se diferentes processos de fabricação ou fornecedores resultam em produtos com médias de qualidade diferentes.

* Educação: Utilizado para avaliar se diferentes métodos de ensino têm um impacto estatisticamente significativo no desempenho dos alunos.

>**Requisitos para Aplicação**

* Grupos Independentes: O ANOVA é adequado quando se deseja comparar três ou mais grupos independentes. Cada grupo representa uma categoria ou tratamento diferente.

* Variável Dependente Métrica: A variável de interesse (variável dependente) deve ser métrica, ou seja, uma variável numérica que pode ser medida quantitativamente.

* Hipóteses e Nível de Significância: O pesquisador deve estabelecer claramente as hipóteses nula e alternativa antes do teste. O nível de significância (geralmente 0,05) também deve ser definido.

* Independência das Observações: As observações dentro de cada grupo devem ser independentes uma das outras, e os grupos não devem ser afetados uns pelos outros.

* Homogeneidade das Variâncias: É importante que a variância das variáveis dependentes seja aproximadamente a mesma em todos os grupos. Violations desse requisito podem ser tratadas com métodos alternativos, como ANOVA com correção de Welch.

>**Interpretação dos Resultados**

* Valor F: O teste ANOVA gera um valor de estatística de teste F, que compara as variações entre os grupos com as variações dentro dos grupos. Um valor F maior indica uma maior probabilidade de diferenças significativas entre os grupos.

* Valor-p: O valor-p é calculado com base no valor F e indica a probabilidade de obter os resultados observados, supondo que não haja diferença significativa entre os grupos. Um valor-p baixo (geralmente menor que 0,05) sugere que pelo menos um dos grupos é estatisticamente diferente dos outros.

Análise Post Hoc: Se o teste ANOVA mostrar diferenças significativas entre os grupos, análises adicionais, como testes post hoc (por exemplo, teste de Tukey), podem ser realizadas para identificar quais grupos são diferentes entre si.
